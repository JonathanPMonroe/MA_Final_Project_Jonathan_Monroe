### Clarifying “Structural Accountability”

Platforms now routinely present systems like Community Notes as “the crowd” speaking. That claim is politically powerful: it lets the platform frame moderation and misinformation labeling as democratic, neutral, and scalable. If the underlying structure is not actually diverse, independent, decentralized, or well-aggregated, then that framing is illegitimate.

I use the term "structural accountability" to describe the degree to which the system’s internal logic satisfies the conditions for collective intelligence: who is allowed to contribute judgment, how consensus emerges among contributors, and how that consensus is converted into platform outcomes.

In this project, I map that structure directly. I identify who is participating in judgment, how agreement forms, and how those judgments are translated into final platform decisions. Through large-scale data mining, I measure those patterns, critique them, and use that critique to surface where the system could be redesigned.

I treat collective intelligence principles—diversity, independence, decentralization, and aggregation quality—as diagnostics of that structure. Each principle is a measurement axis. It tells us how judgment is organized, whose judgment ends up mattering, and how that judgment becomes something the platform treats as authoritative.

By quantifying structural accountability, this project makes the Community Notes system’s legitimacy claim testable. If parts of the system satisfy these collective intelligence conditions, that shows where the platform can defend its governance structure against accusations of systematic bias. If they don’t, that reveals where the Community Notes decision pipeline is unreliable as a misinformation detection mechanism—and, by extension, where its design needs to change.

### Background and Related Work

Collective intelligence research makes a basic but extremely important claim: “the crowd” is only useful under specific structural conditions. Large groups outperform isolated experts when: individuals bring different information and perspectives rather than repeating the same narrative (diversity); people make their judgments independently instead of copying those who spoke first (independence); influence and information are distributed and locally grounded rather than centralized in a small bloc (decentralization); there is a defined way to turn many individual judgments into a single actionable outcome (aggregation). When those conditions are met, aggregated group judgments can be more accurate, more stable, and in some case, more robust to partisan spin than judgments from professional elites (Surowiecki 2005; Lorenz et al. 2011; Almaatouq et al. 2020). They are not met, for example, when early ratings anchor on later ones, or when a small cluster of high-activity participants dominate decisions. The “crowd” can underperform, and even amplify error, as what appears to be consensus could, in reality, be imitation (Lorenz et al. 2011; Surowiecki 2005).

These structural conditions map directly onto the governance problem for misinformation. Platforms now frame certain moderation and correction mechanisms as if they are expressions of the public will. X’s Community Notes system, for example, is explicitly described as “the community providing context,” rather than the platform declaring a post misleading. That framing matters politically, because it recasts content labeling and the assignment of credibility as democratic, neutral, and scalable, rather than top-down editorial enforcement (Gillespie 2018; Matias 2019). But that claim of legitimacy only holds if the underlying process actually behaves like a collective intelligence system— whereby judgment is meaningfully distributed, agreement emerges without herding, and the final “helpful / not helpful” outcome reflects a consistent aggregation of many distinct inputs rather than the preferences of a small gatekeeping layer (Surowiecki 2005; Lorenz et al. 2011; Almaatouq et al. 2020).

Recent empirical work has started to test those claims directly in the context of crowdsourced fact-checking and Community Notes. Studies of platform-scale fact-checking crowds find that large, politically mixed groups of raters can collectively identify misinformation in ways that approximate expert judgments, and can do so at a speed and scale that expert review alone cannot match (Martel et al. 2024). Research on Community Notes in particular has shown two things: first, that attaching a crowdsourced note to a post reduces downstream engagement and resharing of misleading content, suggesting that these notes can suppress diffusion of misinformation (Chuai et al. 2024); and second, that only a subset of misleading posts ever receives such a note, often after most of the spread has already occurred, which limits system-wide impact (Pilarski et al. 2024). Related work has also begun to analyze how Community Notes surfaces notes that draw agreement from ideologically distant raters, suggesting that the platform is not just counting votes but actively privileging “bridging consensus” across political sub-communities as a credibility signal (Kim et al. 2025; Li et al. 2025). This is an explicit form of aggregation logic: the platform is operationalizing not just “how many people agreed,” but “who agreed with whom,” which is an institutional design choice, not a neutral fact.

This project builds on that line of work by treating Community Notes as a decision pipeline—a system that turns individual human judgments into platform-visible outcomes. Instead of asking abstractly whether misinformation is being “handled,” I evaluate what I call structural accountability: the extent to which the internal logic of the system satisfies the core conditions identified in collective intelligence research. Concretely, I measure four diagnostic dimensions that correspond to those conditions. (1) Diversity: whether attention is distributed across different kinds of claims and topics, or concentrated in a narrow band of high-conflict narratives. (2) Independence: whether later raters appear to be making their own calls or simply reproducing early ratings, which would indicate herding pressure. (3) Decentralization: whether judgment is broadly shared across many contributors or effectively concentrated in a small, high-volume stratum that acts as a de facto moderation layer. (4) Aggregation quality: whether the platform’s final status decisions on notes (“helpful,” “needs more ratings,” etc.) track the underlying pattern of ratings in a stable, monotonic, and explainable way, rather than being arbitrary or selectively enforced. Each of these diagnostic axes corresponds to a known requirement for reliable crowd decision-making (Surowiecki 2005; Lorenz et al. 2011; Almaatouq et al. 2020) and to emerging empirical work on crowdsourced misinformation correction at platform scale (Martel et al. 2024; Drolsbach et al. 2024).

The result is that “the crowd decided” stops being a marketing claim and becomes an empirical question with defined failure modes. If the structure behind Community Notes satisfies these collective intelligence conditions, the platform can plausibly defend its governance process against accusations that it is partisan or arbitrary. If it does not, then decisions presented as “crowd-sourced” are in practice concentrated, imitative, or weakly aggregated—and the system is not behaving as a trustworthy misinformation filter at scale (Gillespie 2018; Matias 2019; Drolsbach et al. 2024).


## Exploratory Data Overview

Before applying large-scale data mining methods, this section examines the structure and scale of the Community Notes datasets to establish what forms of analysis are feasible and what the data can reveal about collective judgment on the platform.

The data span 2021–2025 and capture how notes are written, rated, and resolved. Each table serves a specific role in the decision pipeline:

- **`notes.parquet`** – one record per note, containing metadata about the content, author, and time of creation.  
- **`ratings.parquet`** – one record per rater–note interaction, indicating how a participant judged the note’s helpfulness and why.  
- **`noteStatusHistory.parquet`** – timestamped logs of each note’s evaluation state (e.g., *NEEDS_MORE_RATINGS*, *CURRENTLY_RATED_HELPFUL*).  
- **`userEnrollment.parquet`** – background information on rater eligibility and enrollment cohort.

Together, these tables allow reconstruction of how human judgments flow through the Community Notes pipeline—from initial contribution, to crowd evaluation, to platform-level decision.  

Below, we verify dataset scale, record distribution, and basic temporal coverage to establish the foundation for later analyses of diversity, independence, decentralization, and aggregation quality.


```python
from pathlib import Path
from pyspark.sql import SparkSession

spark = SparkSession.builder.getOrCreate()

ProjectRoot = Path.cwd().parent
DataDir = ProjectRoot / "data" / "interim"

RatingsPath = DataDir / "ratings_parquet" / "ratings.parquet"
NotesPath = DataDir / "processed" / "notes.parquet"
StatusPath = DataDir / "processed" / "noteStatusHistory.parquet"
EnrollPath = DataDir / "processed" / "userEnrollment.parquet"

ratings = spark.read.parquet(str(RatingsPath))
notes = spark.read.parquet(str(NotesPath))
status = spark.read.parquet(str(StatusPath))
enroll = spark.read.parquet(str(EnrollPath))

print(f"Ratings count: {ratings.count():,}")
print(f"Notes count: {notes.count():,}")
print(f"Status records: {status.count():,}")
print(f"Enrollment records: {enroll.count():,}")

ratings.printSchema()
```

```python
summary = {
    "Ratings Rows": ratings.count(),
    "Distinct Raters": ratings.select("raterParticipantId").distinct().count(),
    "Distinct Notes Rated": ratings.select("noteId").distinct().count(),
    "Notes Total": notes.count(),
    "Status Entries": status.count(),
    "User Enrollments": enroll.count()
}

for k, v in summary.items():
    print(f"{k:<25}: {v:,}")
```

## Methodological Design

This project mines what I call "collective intelligence signals"—statistical traces of how a crowd produces, evaluates, and converges on shared judgments. Each diagnostic corresponds to one of the classical conditions for collective intelligence—diversity, independence, decentralization, and aggregation—and each is operationalized using the large-scale Community Notes datasets described above. Together, they allow me to quantify what I term "structural accountability": whether the system’s internal logic satisfies the conditions under which crowds make sound, unbiased collective decisions.

The methodological design, therefore, moves from descriptive mapping to diagnostic measurement, applying established data mining techniques to unveil those structural signals from behavioral data.

---

#### Diversity 

To evaluate diversity, I cluster note embeddings to estimate the topical structure of crowd attention. Each note’s text embedding (generated from the “summary” field using all-MiniLM-L6-v2) is projected into a lower-dimensional space via PCA and UMAP, and then clustered using K-Means. The resulting topics capture high-level domains—such as political fact-checks, media hoaxes, or health misinformation—that define the semantic spread of the system’s focus.

This procedure was implemented in the soon-to-be-uploaded topic_clustering.ipynb. Topic prevalence, entropy, and coverage distributions are computed to measure whether attention is dispersed across multiple informational domains or concentrated in a small set of high-salience topics. Topic clusters are also mapped against note helpfulness rates, revealing whether certain subjects dominate collective attention or produce systematic agreement/disagreement patterns.

---

#### Independence

Independence tests whether raters form their own judgments or reproduce those of others. Using the full rating history, I measure rater–rater dependence via pairwise similarity of rating vectors (as implemented in the soon-to-be-uploaded dependence_modeling.ipynb). Locality-sensitive hashing (LSH) and Jaccard similarity sampling were initially used to estimate dependence density across the population, yielding the structural similarity rate (≈3.4%), indicating high independence overall.

For the purpose of this project, I will also estimate independence in a second way by comparing early vs. late rating agreement on each note and computing the correlation between early helpfulness proportions and final outcomes. If later raters consistently replicate earlier votes, independence declines, signaling herding or visibility bias. Stability across sample fractions and temporal intervals provides a robustness check on these findings.

---

#### Decentralization

Decentralization will be measured as a bipartite network between raters and notes. This structure makes visible how judgment flows through the platform—whether it’s dispersed across many contributors or concentrated in a small, recurring subset. Each rater’s degree reflects how many notes they have rated, and each note’s degree shows how many unique raters engaged with it. From these distributions, I compute inequality metrics such as the Gini coefficient and degree entropy to capture whether participation is broad or bottlenecked.

I will then project this bipartite graph onto the rater layer to analyze patterns of co-participation: how often the same raters appear together across notes. Network modularity and connected-component analysis will reveal whether these contributors form localized clusters or remain broadly integrated. If most rating activity clusters around a small, densely connected group, the system is less decentralized than it appears; if connections are widespread and weakly clustered, that indicates a genuinely distributed decision structure.

---

#### Aggregation

Aggregation test whether Community Notes consistently translates many individual judgments into coherent platform decisions. I will model the mapping between note-level rating distributions and final status outcomes (“Helpful,” “Needs More Ratings,” etc.) to evaluate whether those outcomes follow a stable, monotonic relationship with the underlying helpful vote share. Logistic and ordinal regression models will estimate how well collective outcomes can be predicted directly from crowd behavior.

I will also assess whether aggregation behaves uniformly across topics. If some domains systematically achieve “Helpful” designations under weaker consensus, that would indicate that aggregation thresholds are uneven and encode topical bias. To identify these asymmetries, I will compute a topic-level “overperformance” metric—comparing each topic’s helpful rate to the global average—as an indirect indicator of structural preference.

To benchmark aggregation quality, I will use link analysis to see when Community Notes and professional fact-checkers are talking about the same claim. I will extract every outbound URL that appears in Community Notes (for example, links to Snopes, PolitiFact, Reuters Fact Check, AFP Fact Check, etc.). I will treat those linked fact-check articles as an external reference set.

If a note that earns a “Helpful” status links to a third-party fact-check article that explicitly disputes the same claim, that’s evidence that the platform’s aggregation process elevated a correction that is externally verifiable. If a note fails to become “Helpful” despite linking to an established fact-check, or if a note systematically disagrees with a third-party fact check, that’s evidence of drift between what the crowd judges as “helpful” and what professional fact-checkers have already documented.
